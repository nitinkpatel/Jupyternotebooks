```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
apple_weight = np.random.normal(125,12,1000)
apple_volume = np.random.normal(175,17,1000)
orange_weight = np.random.normal(215,22,1000)
orange_volume = np.random.normal(218,22,1000)
a = {"Weight": apple_weight,"Volume": apple_volume, "Type": np.full(1000,"Apple")}
o = {"Weight": orange_weight,"Volume": orange_volume, "Type": np.full(1000,"Orange")}
```

```python
fruit = pd.DataFrame(a).append(pd.DataFrame(o))
```

```python
sp = sb.scatterplot(x="Volume",y="Weight",data=fruit,hue="Type")
fig = sp.get_figure()
fig.savefig("apples_oranges.png",dpi=600)
```

```python
fruit = fruit.append(pd.DataFrame({"Weight": [125],"Volume": [175],"Type": ["Sample"]}))
fruit = fruit.append(pd.DataFrame({"Weight": [215],"Volume": [218],"Type": ["Sample"]}))
fruit = fruit.append(pd.DataFrame({"Weight": [170],"Volume": [197],"Type": ["Sample"]}))
```

```python
sp = sb.scatterplot(x="Volume",y="Weight",data=fruit,hue="Type")
fig = sp.get_figure()
fig.savefig("apples_oranges_sample.png",dpi=600)
```
